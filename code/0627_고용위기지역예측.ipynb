## 조건 1.
고용보험 피보험자(이하 "피보험자"라 한다) 증감률(고용위기지역 신청 직전 1년간 평균 피보험자 수와 해당 기간 전 1년간 평균 피보험자 수를 토대로 산정한 증감률을 말한다)이 같은 기간 전국 평균 피보험자 증감률 보다 5%p 이상 낮은 경우

```python
df_insurance_final = df_insurance_final.apply(pd.to_numeric, errors='coerce')
df_insurance_final = df_insurance_final.fillna(0)
```

```python
df_insurance_final['전국 총합'] = df_insurance_final.sum(axis=1)
```

```python
current_period = df_insurance_final['2024-04-01':'2025-03-01']
previous_period = df_insurance_final['2023-04-01':'2024-03-31']
```

```python
# 직전 1년과 그 전 1년 동안의 평균 피보험자 수 계산
current_avg = current_period.mean()
previous_avg = previous_period.mean()
```

```python
# 증감률 계산 (백분율)
growth_rate = ((current_avg - previous_avg) / previous_avg) * 100
```

```python
# 전국 평균 피보험자 수 증감률 계산
national_current_avg = current_period['전국 총합'].mean()
national_previous_avg = previous_period['전국 총합'].mean()
national_growth_rate = ((national_current_avg - national_previous_avg) / national_previous_avg) * 100
```

```python
# 전국 평균 피보험자 증감률보다 5%p 이상 낮은 경우 필터링
threshold = national_growth_rate - 5
below_threshold = growth_rate[growth_rate < threshold]
```

```python
# 전국 평균 피보험자 증감률보다 3%p 이상 낮은 경우 필터링
threshold3 = national_growth_rate - 3
below_threshold3 = growth_rate[growth_rate < threshold3]
```

```python
# 전국 평균 피보험자 증감률보다 1%p 이상 낮은 경우 필터링
threshold1 = national_growth_rate - 1
below_threshold1 = growth_rate[growth_rate < threshold1]
```

```python
# 전체 지역 증감률을 데이터프레임으로 저장
overall_growth_rate_df = pd.DataFrame({
    '지역': growth_rate.index,
    '증감률': growth_rate.values
}).reset_index(drop=True)
```

```python
overall_growth_rate_df
```

```python
filtered_df = overall_growth_rate_df[overall_growth_rate_df['지역'] == '경상남도 고성군']
filtered_df
```

```python
# 5%p 이상 낮은 지역을 데이터프레임으로 저장
below_threshold5_df = pd.DataFrame({
    '지역': below_threshold.index,
    '증감률': below_threshold.values,
    '전국 평균 증감률': national_growth_rate
}).reset_index(drop=True)
```

```python
# 3%p 이상 낮은 지역을 데이터프레임으로 저장
below_threshold3_df = pd.DataFrame({
    '지역': below_threshold3.index,
    '증감률': below_threshold3.values,
    '전국 평균 증감률': national_growth_rate
}).reset_index(drop=True)
```

```python
# 1%p 이상 낮은 지역을 데이터프레임으로 저장
below_threshold1_df = pd.DataFrame({
    '지역': below_threshold1.index,
    '증감률': below_threshold1.values,
    '전국 평균 증감률': national_growth_rate
}).reset_index(drop=True)
```

## 조건 3.
고용위기지역 신청 직전 1년간 구직급여 신규 신청자 수가 해당 기간 전 1년간 구직급여 신규 신청자 수보다 100분의 20 이상 증가한 경우

```python
# 비교 기간 설정
period_1_start = '2023-04-01'
period_1_end = '2024-03-01'
period_2_start = '2024-04-01'
period_2_end = '2025-03-01'
```

```python
# 각 지역별로 구직급여 신규 신청자 수 계산
results = []
results_2 = []
results_3 = []
```

```python
for region in df_jobs_final.columns:
    # 첫 번째 기간의 데이터
    period_1_data = df_jobs_final.loc[period_1_start:period_1_end, region]
    period_1_sum = period_1_data.sum()

    # 두 번째 기간의 데이터
    period_2_data = df_jobs_final.loc[period_2_start:period_2_end, region]
    period_2_sum = period_2_data.sum()

    # 증가율 계산
    if period_1_sum != 0:  # 0으로 나누는 것을 방지
        increase_rate = ((period_2_sum - period_1_sum) / period_1_sum) * 100
        if increase_rate >= 20:
            results.append((region, increase_rate))
```

```python
for region in df_jobs_final.columns:
    # 첫 번째 기간의 데이터
    period_1_data = df_jobs_final.loc[period_1_start:period_1_end, region]
    period_1_sum = period_1_data.sum()

    # 두 번째 기간의 데이터
    period_2_data = df_jobs_final.loc[period_2_start:period_2_end, region]
    period_2_sum = period_2_data.sum()

    # 증가율 계산
    if period_1_sum != 0:  # 0으로 나누는 것을 방지
        increase_rate_2 = ((period_2_sum - period_1_sum) / period_1_sum) * 100
        if increase_rate_2 >= 15:
            results_2.append((region, increase_rate_2))
```

```python
for region in df_jobs_final.columns:
    # 첫 번째 기간의 데이터
    period_1_data = df_jobs_final.loc[period_1_start:period_1_end, region]
    period_1_sum = period_1_data.sum()

    # 두 번째 기간의 데이터
    period_2_data = df_jobs_final.loc[period_2_start:period_2_end, region]
    period_2_sum = period_2_data.sum()

    # 증가율 계산
    if period_1_sum != 0:  # 0으로 나누는 것을 방지
        increase_rate_3 = ((period_2_sum - period_1_sum) / period_1_sum) * 100
        if increase_rate_3 >= 13:
            results_3.append((region, increase_rate_3))
```
